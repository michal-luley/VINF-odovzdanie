```python
import os
import sys
import pickle
import pandas as pd
```

```python
# Setting up home directory
D_RAW_DIR = os.path.realpath(os.path.join(os.path.dirname("VINF"), '..'))
```

```python
# Specifying the path to files 
main_document_path = os.path.realpath(os.path.join(D_RAW_DIR, 'source_files', 'main_document.csv'))
symptom_index_path = os.path.realpath(os.path.join(D_RAW_DIR, 'indexes', 'symptoms_index.pickle'))
type_index_path = os.path.realpath(os.path.join(D_RAW_DIR, 'indexes', 'type_index.pickle'))
country_index_path = os.path.realpath(os.path.join(D_RAW_DIR, 'indexes', 'country_index.pickle'))
transmission_index_path = os.path.realpath(os.path.join(D_RAW_DIR, 'indexes', 'transmission_index.pickle'))

# Loading main file
main_df = pd.read_csv(main_document_path, sep = '\t')
main_df = main_df.fillna("")
```

```python
# Main interaction
while True:
    # Category selection
    mode = int(input(f"\nVyber podľa čoho hľadať ochorenie(cislo): \n 1. typ ochorenia\n 2. symptomy ochorenia\n 3. krajina ochorenia\n 4. sposob prenosu ochorenia\n 5. ukoncit\n"))
    while mode not in [1, 2, 3, 4, 5]:
        print("Nespravny udaj\n-------------------")
        mode = input("Zadaj cislo (1,2,3,4,5): ") 

    if mode == 1:
        index_file = type_index_path
    elif mode == 2:
        index_file = symptom_index_path
    elif mode == 3:
        index_file = country_index_path
    elif mode == 4:
        index_file = transmission_index_path
    else:
        print("\n--------Dovidenia---------\n")
        break

    # Printing all available items to choose from
    index_dict = {}
    with open(index_file, 'rb') as handle:
        index_dict = pickle.load(handle)

    for item in index_dict.keys():
        print(item)

    # Choosing item to search diseases for
    option = input("------------------\nVyber hodnotu: ")
    while index_dict.get(option) is None:
        print("Zly vstup")
        option = input("Vyber hodnotu: ")

    print("Zodpovedajuce ochorenia:\n")
    for item in index_dict.get(option):
        print(f"         {main_df['Disease'][item]}")    
          
```
